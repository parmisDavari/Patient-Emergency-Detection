3.3. Outlier Analysis

```python
##################################################################################################################
def outlier_thresholds(dataframe, variable):
    quartile1 = dataframe[variable].quantile(0.05)
    quartile3 = dataframe[variable].quantile(0.95)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit
##################################################################################################################
def replace_with_thresholds(dataframe, variable):
    low_limit, up_limit = outlier_thresholds(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit
##################################################################################################################
def has_outliers(dataframe, num_col_names, plot=False):
    variable_names = []
    for col in num_col_names:
        if dataframe[col].dtype.name != 'category':  # Check if the column is not categorical
            low_limit, up_limit = outlier_thresholds(dataframe, col)
            outliers = dataframe[(dataframe[col] > up_limit) | (dataframe[col] < low_limit)] 
            number_of_outliers = outliers.shape[0]
            if number_of_outliers > 0:
                print(col, ":", number_of_outliers)
                variable_names.append(col)
                if plot:
                    sns.boxplot(x=dataframe[col])
                    plt.show()
    return variable_names
```

5.1. One Hot Encoding and Robust Scaler

```python
def one_hot_encoder(dataframe, categorical_columns, nan_as_category=False):
    original_columns = list(dataframe.columns)
    dataframe = pd.get_dummies(dataframe, columns=categorical_columns,
                               dummy_na=nan_as_category, drop_first=True)
    new_columns = [col for col in dataframe.columns if col not in original_columns]
    return dataframe, new_columns

def robust_scaler(variable):
    var_median = variable.median()
    quartile1 = variable.quantile(0.05)
    quartile3 = variable.quantile(0.95)
    interquantile_range = quartile3 - quartile1
    if int(interquantile_range) == 0:
        quartile1 = variable.quantile(0.05) 
        quartile3 = variable.quantile(0.95)
        interquantile_range = quartile3 - quartile1
        z = (variable - var_median) / interquantile_range
        return round(z, 3)
    else:
        z = (variable - var_median) / interquantile_range
    return round(z, 3)
```

1. like_num = [col for col in new_df.columns if new_df[col].dtypes != 'O' and len(new_df[col].value_counts()) > 10]:
This line creates a list called like_num.
It iterates through each column (col) in the DataFrame new_df.
For each column, it checks two conditions:
new_df[col].dtypes != 'O': This condition ensures that the data type of the column is not an object (i.e., not a string or categorical data).
len(new_df[col].value_counts()) > 10: This condition checks if the number of unique values in the column (determined by value_counts()) is greater than 10.
If both conditions are met, the column name (col) is added to the like_num list.
2. cols_need_scale = [col for col in new_df.columns if col not in like_num and col not in 'mistriage']:
This line creates another list called cols_need_scale.
It iterates through each column (col) in the DataFrame new_df.
For each column, it checks two conditions:
col not in like_num: This ensures that the column is not already included in the like_num list (i.e., it’s not a numeric column with more than 10 unique values).
col not in 'mistriage': This condition excludes columns named ‘m’, ‘i’, ‘s’, ‘t’, ‘r’, ‘i’, ‘a’, ‘g’, ‘e’ (which is likely a mistake; it should probably be a list of column names).
If both conditions are met, the column name (col) is added to the cols_need_scale list.
like_num contains the names of numeric columns in new_df with more than 10 unique values.
cols_need_scale contains the names of columns (excluding those in like_num and any named ‘mistriage’) that need scaling or further processing.
